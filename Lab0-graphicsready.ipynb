```python
## Read some data 
## The Source of these data are https://www.census.gov/population/international/ 
pop=Table().read_table(datasite+"histpop.csv")

pop.show()


```

```python
#Plot the table we just read
plt.plot(pop['yearsSinceNow'],pop['popLower'])
plt.title("World Population Since a Long Time Ago")
plt.ylabel("Millions of Humans")
plt.xlabel("Years Since Present")
plt.show()
#pop.plot(select=['popLower','popUpper'],column_for_xticks='yearsSinceNow')
```

```python
#Add another line to the chart to show upper range of estimates

plt.plot(pop['yearsSinceNow'],pop['popLower'],pop['popUpper'])

plt.title("World Population Since a Long Time Ago")
plt.ylabel("Millions of Humans")
plt.xlabel("Years Since Present")
plt.show()
#pop.plot(select=['popLower','popUpper'],column_for_xticks='yearsSinceNow')
```
