Let's review for a moment how we find the exponential growth rate $r$ given to populations T years apart:

$$ P_{T} = P_{0} * e^{rT}$$ 

or in other words:

$$ log(\frac{P_{T}}{P_0}) = rT $$

or more succinctly:

$$ r = \frac{log(\frac{P_{T}}{P_0})}{T}$$

# The logistic growth model

So the constant exponential growth model is maybe not so good for human populations -- at least over long time periods. But at the same time, the model does hold some promise.  Recall that it follows quite logically from the assumptions that the birth rate and the death rate are constant proportions of the current population.  What happens if we relax that assumption   and allow $(B/P - D/P)$ to change over time?

Well, we could simply compute a separate exponential growth rate for each period. That's not hard to do with historical data, and it results in a good description of a population's historical path. It doesn't however, do much for predicting tomorrow's population. 


## The logistic model is a model of the change in exponential growth rates.

Consider a population of simple organisms with a large but finite amount of space; a constant supply of food; and no predators. Mold growing in your disgusting roommate's abandoned latte is an example of this sort of situation.

What we might expect in this scenario is for population to grow rapidly for a while at some maximum exponential rate ... and then to slow down and level off as "carrying capacity" is reached.  It's an S-shaped population trajectory such as we might find in the graph below.  Note that the graph below is of something entirely different.

### The logistic model  and Malthus ?

##### Before we turn to the details of the logistic growth model -- let's consider Malthus's description of the world. Is there something in Malthus chapters that you read that might favor an S-shaped population growth? 


### The details

### (Eq 3): $\frac{dP}{dt}=r_{max}(\frac{K-P}{K})P$

Equation 3 is technically a differential equation -- This is NOT a math class you do not have to know what a differential equation is. It's just easier to  understood the logistic model in this form.  On the left side we have $\frac{dP}{dt}$ which is the derivative of Population, P with respect to time, t. It is perfectly reasonable to think of this "derivative" as a rate of change or in other words the rate at which the population changes over an infinitessimally small time period (that should sound familiar). If you prefer not to think infinitesimally, then thing of $dP/dt$ as the *change* in P/ *change* in t or in symbols, $\Delta P / \Delta t$ 

For comparison, if we were to write the exponential growth model as a differential equation it would look like this:

### (Eq 4): $\frac{dP}{dt} = rP$



The left sides of Eq 3 and Eq 4 are the same, but in place of all the cools stuff on the right side of Eq3 we have instead simply the constant rate $r$ multiplied by the population at that moment.  Looking a little closer we can see that in Eq3 we have the fancy expression $r_{max}(\frac{K-P}{K})$ taking the place of the simple constant $r$ in Eq4.

Think of $r_{max}$ as the "maximum" exponential growth rate possible and think of the expression $\frac{K-P}{K}$ as a factor which falls as P rises.  Think of K as the "carrying capacity" or the highest population that can be sustained with the available resources.  When P is close to zero, $\frac{K-P}{K}$ is close to 1 and the logistic differential equation reduces to something quite close to the exponential differential equation in Eq4. So when the population is small relative to carrying capacity, it grows at a rate close to $r_{max}$. On the other hand,  when P is close to K, $\frac{K-P}{K}$ approaches zero and that implies that the entire right hand side of Eq3 goes to zero and population growth stops.

Put that all together and we have a population that starts out growing fast $(P<<K)$ then as P rises towards K, the growth *rate* slows until ultimately approaching zero as P approaches K.

## Optimization 

One of the weaknesses of our comparison of the logistic model with the observed population data is that we assume that K, the carrying capacity is equal to the maximum observed population.  That follows from the story that I told you about hunter gathers and agriculturalists -- wherein new technology was only adopted when the world could no longer support the population under the current technology.

Of course, that was an oversimplification and highly dependent on the context.  In other contexts, one might want to consider K to be unknown.  In fact, one might like to use data to predict K -- since, under the logistic model, K is where the population is headed in the long run.


In the cells below,  we'll take this approach with the postindustrial world -- where population has not stopped growing just yet  and with the 19th Century US population, a classic example from Pierre Verhulst, who is credited with inventing (discovering?) the logistic equation.

In data 8, you'll get to optimization in the regression context in several weeks.  The treatment below is a bit simpler (cruder) than what you'll see later in data 8 -- but hopefully that will make it easier to understand.

In a nutshell, we're going try a bunch of possible values of K and pick the best one.  By "best" we mean the value that produces the smallest absolute difference between the populations that are implied by that particular value of K  and those observed in the real world.  

More specifically,  we're going to assume that we know r_max to be 0.041 because that is the fastest human population growth yet observed.  Then we'll compute population trajectories based on the assumed r_max, the observed starting population and the trial value of K.  We'll compare the hypothetical trajectories to the observed populations by subtracting one from the other; taking the absolute values and then adding all the absolute differences together.  That will yield, for each value of K, a positive number.  The K with the smallest associated sum of absolute differences is the best fitting K.  the best fitting K is our forecast of the maximum population under the logistic model.

We'll do the above in three steps:

1. We'll create a function that for a given set of observed population figures and a given K returns the absolute sum of differences between the observed populations and those implied by the given K (along with all the other assumptions discussed above)
2. We'll run the function giving it a set of evenly spaced K values over a range and identify the "best" K value.
3. We'll run the function over a new set of K's which are more tightly spaced around the K value identified in step 2

We could repeat steps 2 and 3 if we want to get really precise,  but given the limitations of our assumptions ... that could be a waste of electrons.


```python
## Fitting logistic 

## this the first step  -- a function to assess absolute sum of differences between observed and expected 
## populations given K & etc.

def fitlogistic(Xpop,K):
    ## we know r_max and we observe an epoch of population
    ## for a given K what's the absolute error in deltaPs  that is what is the 
    ## SUM of ablsolute values of the differences between the observed deltaP (computed from data)
    ## and the expected deltaP computed from the logistic differential equation formula with specified
    ## values of r_max, P and K
    
    #  Observed delta P  -- add this column to epoch so we can evaluate fit
    deltaPobs=list(np.diff(Xpop['popLower'])/np.diff(Xpop['time']))
    deltaPobs.append(None)
    Xpop.append_column('deltaPobs',deltaPobs)
    
    P0=Xpop['popLower'][0]    #initial population 
    
    r_max=.041  # maximum achievable growth rate .041 is max observed human pop growth. this ignores
    # net migration but it still exceeds the fastest growth experienced by US
    
    # variables to hold results
    P=[]  # computed population
    deltaPlogistic=[]  # deltaP from logistic model
    Plast=P0  # population in previous period - needed in loop for computation
    time=np.arange(Xpop['time'][-1])  # period over which we compute logistic population model
    for t in time:
        deltaP=r_max*((K-Plast)/K)*Plast
        P.append(Plast+deltaP)
        Plast+=deltaP
        deltaPlogistic.append(deltaP)
    
    ## put results in a table so that it can be joined with the observed data
    tab3=Table().with_columns('time',time,'P',P,'deltaPlogistic',deltaPlogistic)
    ##  the join operation merges rows in two tables by matching on values of one variable
    ##  in this case time.  The resulting table ONLY holds data where there are equivalent 
    ##  time values in both input tables.  many rows from tab3 will be lost in this join
    tabRes=Xpop.join('time',tab3,)

    res=np.sum(np.abs(tabRes['deltaPlogistic']-tabRes['deltaPobs']))
    return(res)
```

```python
## Optimization 

# this cell is for the second and third step -- choosing the best K and refining the search

# Let's use the fitlogistic function to search over reasonable values of K for the one that
# fits the data best.  The combination of r_max (which we assume to be 0.041) and a value of K
# define a particular logistic curve -- we choose the one that produces estimates of delta P that
# are closest to those observed

EPOCH=postindustrial  # try agriculture, industrial, postindustrial
epoch=pop.where('yearsSinceNow', are.between_or_equal_to(EPOCH[0],EPOCH[1]))
epoch.append_column('time',(epoch["yearsSinceNow"]-epoch["yearsSinceNow"][0]))



## np.arange  determines which values K (khat) will be evaluated
## khat is defined as the final population observed multiplied by x (which is set by np.aranage)
## so if x= .7 then khat would be 70% of the final observed population if x=2 khat will be
## twice the final observed population.  Step size determines how many Ks will be tried.
for x in np.arange(.5,3,step=.05):
    khat=(epoch['popLower'][-1])*x
    fit=fitlogistic(epoch,K=khat)
    print("x={0} khat={1} fit={2}".format(x,khat,fit))

    
##  Noting that 1.25 was the optimum from the first seach... we'll refine our set of Ks
## to search over

for x in np.arange(1.2,1.3,step=.005):
    khat=(epoch['popLower'][-1])*x
    fit=fitlogistic(epoch,K=khat)
    print("x={0} khat={1} fit={2}".format(x,khat,fit))
  
```
