##  Key ideas:

1. Timing of reductions in mortality then fertility produces:
    1. A permanent-ish  population increase.  The increase in population is permanent to the extent that fertility stabilizes at around replacement levels.  While it is frequently assumed that population eventually tend to get there... there is no guarantee.  Many palces in the world e.g. Germany, China, Japan, Italy, South Korea, Estonia  have and continue to have sub replacement fertilities for very long times.
    
    1. A temporary change in the population age structure. When fertility finally falls, during the demographic transition,  the number of dependent children relative to working age persons falls. This provides an opportunity for the economy to grow as family incomes rise while costs (to support children) fall.  Welfare and investment may rise.  *BUT* eventually fertility and mortality come back into balance and the surplus of working age people disappears or turns negative as the proportion of old post working age population rises. 
    
    A "second" demographic dividend or a different type is *possible* if older workers face strong incentives to save for an extended retirement. If, in the absence of children or state support, older workers save (as has been observed in some countries)  that can generate important increases in capital stock leading to sustained growth even after the population bulge passes.  This second dividend is both less purely demographic and more speculative, but the key point is that the effects on ages structure are transient -- the the demographic dividend(s) are  opportunities which require sound economic policies if they are turn into growth.
    
###  If economic policies are inadequate or technology is too rigid then the bulge in the prime working age population can turn into a wave of emigration.

##  Crude rates:  birth, death and migration

A useful way to look at population data over time is with so called "crude" rates.  Demographers call these rates "crude"  because they do not take into account the age structure of a population. Crudeness limits their usefulness in some settings, but for long historical series,  it's often the best we have.  AND more to the point, crude rates are great for seeing the <b>demographic transition</b>.

The Crude birth rate (CBR) refer to a period of time, in the present case five year intervals, but they are often - as with the rates we will soon look at - presented as *annual rates*  This does some violence to both precision and accuracy, but in most case we can live with it.

Crude birth rate:  CBR gives births during the time period divided by the person years lived by the population as a whole.

Crude death rates: CDR gives deaths during the time period divided by the person years lived by the population as a whole.

Net Migration Rate: NetMigR give the *net* number of people who have immigrated during the time period  divided by the person years lived by the population as a whole.

#### Happily the UN tabulates these crude rates for all countries since 1950 and forecasts them to 2100

## Population growth rate 

The table that we just read does not include the population growth rate.

Let's define the "rate of natural increase" (RNI) to be the $\frac{\text{number of births - number of deaths}}{\text{midyear population}}$  In other words proportion by which the population grows if we ignore immigration and emigration.

What is RNI in terms of CBR and CBD?

## Crude rates neglect the age distributions

We have spent a lot of time in previous labs trying to deal with the effects of age on mortality and fertility. It would be cowardly to simply ignore age at this point,  but in the context of the demographic transition,  changes in the population age distribution can be seen more as an effect than as a cause.

Let's now take a look at the age distributions in the context of the historical crude rates.

Data on age distributions take up much more RAM than the crude rates table because for each year X country we need the number of people at each age.  Consequently we'll read them differently. Rather than reading one large file as we did for the crude rates, we'll read only the data for the particular country that we are focusing on.

We can do this because each country's data lives in a separate file.  We'll first read a list of country files to see what is available. Armed with that table we'll be able to avoid ever learning how to spell Kyrgystan

```python
# country data files

#  read in a table of names of country data files that we can read
cts=Table().read_table(datasite+"countries.csv")
## construct an np.array of country file names so that in the rest of this lab
## we'll be able to specify Kyrgystan as countries[107] and thus never learn how to spell it.
countries=np.array(cts['x'])
#countries.show()
Table().with_columns('country name',countries,'index',np.arange(len(countries))).show()
```

```python
## Let's look at population age distribution of the whole world and how it's changed over time
##

plt.plot(world.where('year',1950)['age'],world.where('year',1950)['N'],'b-')
plt.plot(world.where('year',2015)['age'],world.where('year',2015)['N'],'y:')
plt.title("World population by age 1950 v 2015")

plt.show()

plt.plot(world.where('year',1950)['age'],world.where('year',1950)['N']/np.sum(world.where('year',1950)['N']),'b-',
         label='1950')
plt.plot(world.where('year',2015)['age'],world.where('year',2015)['N']/np.sum(world.where('year',2015)['N']),'y:',
        label='2015')
#handles,labels = plt.get_legend_handles_labels()

plt.title("World population age structure: 1950 v 2015 ")
plt.legend()
plt.show()
```

```python
## generalize the above into a function
def ageDist(cnt,years=[1950,2015]) :
    """
    uploads data for a country and plots age distribution for specified years
    """
    lwd=1
    print("uploading data on "+cnt)
    titlstr=cnt +' population age distributions'
    pop=Table().read_table(datasite+cnt)
    for yr in years :
        plt.plot(pop.where('year',yr)['age'],pop.where('year',yr)['N']/np.sum(pop.where('year',yr)['N']),
                 linewidth=lwd, label=str(yr))
        titlstr+=str(yr)+' '
        lwd+=1
    plt.title(titlstr)
    plt.legend()
    plt.show()
```

## Dependency Ratios

Comparing age distributions among populations is difficult because there so many numbers involved, a good statistic is one that is (1) easy to compare across times and populations and (2) which measures something meaningful.  In the context of the demographic transition and it's hoped for demographic divident, dependency ratios meet both of these goals

DRs  show the number of old and/or young people in the population per person of working age.  Low DRs  imply a large labor supply relative to the number of dependents; high  DRs imply some combination of large numbers of old and/or young people in the population relative to those of working age.

Computing them is straight forward, but a little involved.  As usual, we'll do the computation once (for China)  then we'll consolidate the code into a function that we can call for all sorts of other countries.



```python
# dependancy ratios
## first a convenient function to read the data for a country
def getCountry(cnt):
    pop=Table().read_table(datasite+cnt)
    return(pop)
```

```python
pop=getCountry('China.csv')
young=pop.where('age',are.between_or_equal_to(0,15)).groups(['country','year'],collect=np.nansum)
middle=pop.where('age',are.between_or_equal_to(16,64)).groups(['country','year'],collect=np.nansum)
old=pop.where('age',are.between_or_equal_to(65,100)).groups(['country','year'],collect=np.nansum)
              
young.relabel('N nansum','Nyoung')
middle.relabel('N nansum','Nmiddle')
old.relabel('N nansum', 'Nold')
```

```python
depRatio=young.join('year',middle).join('year',old)
depRatio.show(5)
depRatio=depRatio.select(['year','country','Nyoung','Nmiddle','Nold'])
depRatio
```

```python
depRatio.append_column('oldDR',depRatio['Nold']/depRatio['Nmiddle'])
depRatio.append_column('youngDR',depRatio['Nyoung']/depRatio['Nmiddle'])
depRatio.append_column('totDR',(depRatio['Nold']+depRatio['Nyoung'])/depRatio['Nmiddle'])
depRatio.append_column('wrkpct',depRatio['Nmiddle']/(depRatio['Nold']+depRatio['Nyoung']+depRatio['Nmiddle']))
```

```python
plt.subplot(2,1,1)
plt.plot(depRatio['year'],depRatio['oldDR'], label="old DR")
plt.plot(depRatio['year'],depRatio['youngDR'], label="young DR")
plt.plot(depRatio['year'],depRatio['totDR'], label="Total DR")
plt.legend()
plt.subplot(2,1,2)
plt.plot(depRatio['year'],depRatio['wrkpct'])
plt.title("Percent Working Age")
plt.ylabel("pct of pop working age")
plt.show()
```

```python
## As usual, let's make a function out this 
## we copy all the code we developed above and make it general by substituting 'cnty'
## for the specific country name
def plot_depRatio(cnty):
    """
    downloads data for cnty computes and plots dependency ratios
    """
    print(cnty+" dependency ratio")
    pop=getCountry(cnty)
    young=pop.where('age',are.between_or_equal_to(0,15)).groups(['country','year'],collect=np.nansum)
    middle=pop.where('age',are.between_or_equal_to(16,64)).groups(['country','year'],collect=np.nansum)
    old=pop.where('age',are.between_or_equal_to(65,100)).groups(['country','year'],collect=np.nansum)
              
    young.relabel('N nansum','Nyoung')
    middle.relabel('N nansum','Nmiddle')
    old.relabel('N nansum', 'Nold')
    
    depRatio=young.join('year',middle).join('year',old)
    depRatio=depRatio.select(['year','country','Nyoung','Nmiddle','Nold'])
    
    depRatio.append_column('oldDR',depRatio['Nold']/depRatio['Nmiddle'])
    depRatio.append_column('youngDR',depRatio['Nyoung']/depRatio['Nmiddle'])
    depRatio.append_column('totDR',(depRatio['Nold']+depRatio['Nyoung'])/
                           depRatio['Nmiddle'])
    depRatio.append_column('wrkpct',depRatio['Nmiddle']/
                           (depRatio['Nold']+depRatio['Nyoung']+depRatio['Nmiddle']))
    
    
    plt.subplot(2,1,1)
    plt.plot(depRatio['year'],depRatio['oldDR'], label="old DR")
    plt.plot(depRatio['year'],depRatio['youngDR'], label="young DR")
    plt.plot(depRatio['year'],depRatio['totDR'], label="Total DR")
    plt.legend()
    plt.subplot(2,1,2)
    plt.plot(depRatio['year'],depRatio['wrkpct'])
    plt.title("Percent Working Age")
    plt.ylabel("pct of pop working age")
    plt.subplots_adjust(hspace=.5)
    plt.show()


    # we don't need to return depRatio because the function does the plots that we want
    # but if you want to work with it just uncomment the next line
    # return(depRatio)
```
