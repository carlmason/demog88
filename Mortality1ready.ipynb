## swe1751  is called a 'cohort life table'

It contains everything demographers generally know about the mortality experience of the 'cohort' of Swedes who were born in 1751.  We're using Swedish data here because, unlike every other country on earth, the Swedes collected this data way back then.  Despite being a long time ago, 1751 is an interesting year because right about this time, the English are starting to mechanize cloth production and thereby launching ...the Industrial Revolution -- which is and was a very big deal both demographically and in every other way.

The structure of the life table reflects its ancient origins  (the first was published in 1662 by John Graunt) in that three of the columns are interesting and the rest are just intermediate calculations -- they are handy for various other obscure things that demographers do but mostly they are in the table so that a person without a modern computer can construct the $e_x$ column, described below.

### The column named 'lx' (lower case L)  is the starting point.  
It contains the number of people who survive to the beginning of each Age.  In the top row, Age =0 refers to birth -- the beginning of year of age 0.  The lx column shows 100000 -- How convenient, you might think, that such a round number of Swedes happened to be born in 1751-4.   But no -- All the numbers in the lx column have been multiplied by the $\frac{100,000}{actual number of births}$  Often one divides the lx column by that 1000000 which gives us the proportion of people born who survive to the beginning of each year of Age.

"Normalizing" the cohort size,  makes the lx curve easier to work with (important with the computing power available in 1661) -- also easier to compare across years -- as you see in the graphs in the next cell.

## The 'qx' and 'ex' columns show *almost* the same information but in enlighteningly different ways.



#### The 'qx' column gives the probabilty of death between age x and x+1 
$q_x = \frac{\text{nuber who die between age x and x+1}}{number who survive to age x}$ or more succinctly
$q_x = \frac{l_{x}-l_{x+1}}{l_x}$
####  The 'ex' column gives the years of expected life remaining at the beginning of age x
We're going to compute $e_x$ so we'll save the details for later.

Each of the the three important columns have characteristic shapes:

```python
swe1751.plot('Age',select='lx')
swe1751.plot('Age',select='qx')
swe1751.plot('Age',select='ex')
```



## The column in which *most people are most interested* is $e_x$ the expected years of remaining life at each age  aka  *life expectancy*.
The number in the first row (Age = 0) of the $e_x$ column is the expected remaining years of life at birth -- or more commonly the "life expectancy". Demographer refer to it affectionately as "$e_0$.  It requires  a few calculations to derive $e_x$ from $l_x$ the traditional way -- that you'll learn about when you minor in demography. We're going to take a more 21st Century data sciency approach which uses the tools we already have.  It also makes the meaning of $e_x$ a little clearer... I think.

Instead of 18th Century algebra, we're going generate a simulated population with death ages corresponding to $lx$.  From our simulated population or list of life spans, we can compute $e_x$ AND all sorts of interesting values which characterize what life (and death) was like over the centuries.

To simulate ages at death, we'll work with a quantity derivable from the $l_x$ : 

### The unconditional  probability of death at any age x  
$pdeath_x = \frac{l_{x}-l_{x+1}}{l_0}$

This is very similar to the formula for $q_x$ in terms of $l_x$ but the denominator is different.

In the next cell we'll create a function that computes 'pdeath' for each age in a life table.  We'll append it as an additional column because it's easiest to keep track of that way.


```python
## adding a colum to a life table
def pdeathF (ltab):
    """
    Adds a column to a life table the probability of dying during each Age conditional only on being born.  So 
    unconditional really.
    """
    ## np.diff(x) returns an array constructed by subtracting the nth element of x from the n+1 th element
    ## since lx is decreasing the result is negative but its the negative of the number of deaths that ocurred
    ## during each age interval
    ## take absolute value to flip the sign
    
    pdeath= abs(np.diff(ltab['lx'])/ltab['lx'][0])
       
    # generally the prob of surviving top age is zero but it doesn't have to be
    ltab.append_column('pdeath',np.append(pdeath,1-pdeath.sum()))
    return(ltab)
    
```

```python
## Using the pdeathF  function ...
swe1751=pdeathF(swe1751)
swe1751
swe1751.plot('Age', select="pdeath")

##note that pdeath and qx are the same in row Age 0  but not so much thereafter
```

### Now let's use pdeath to compute $e_0$ for the 1751 birth cohort

##### The reason we can use pdeath to estimate $e_0$ is because $e_0$, the life expectancy at birth is the same thing as the average age at death.



In the cell below we use the .sample method to draw a sample (with replacement) of rows of the swe1751 life table. The clever part is that we'll use the new 'pdeath' column as weights.  This means that our resulting sample will have lots of copies of the rows in the life table -- but it will have more copies of rows for which 'pdeath' is higher.  If we let each row in the sample represent a simulated person, then we can take their 'Age' as their age at death -- or more precisely their age at their last birthday before death.  We can improve our estimate of the age at death by adding the $a_x$ column to the Age column. The $a_x$ column, as luck would have it, is the average day of death measured as a proportion of a year. 

```python
## simulating ages at death

## here we are taking the weighted sample
pop1751=swe1751.sample(k=100000,with_replacement=True,weights=swe1751['pdeath'])
## now we just take the mean
e0_good=np.mean(pop1751['Age'])
e0_better=np.mean(pop1751['Age']+pop1751['ax'])
e0_hmd=swe1751.where('Age',0)['ex']
print("""
e0 ignoring years of life lived during the age interval of death:{0}
e0 using 'ax' which is the estimated number of years lived 
during the age interval by those who die in it:{1}
e0 from the human mortality database:{2}

Run this cell again ...""".format(e0_good,e0_better,e0_hmd))
```

##  HOW Is it possible to compute a life expectency for 2017 when no one born in 1917 in  has reached the age of 2 yet?



The answer is at once complicated and yet simple. Life tables begin with the lx column. In the case of cohort, this is quite natural because one starts with a certain number of people born in a certain year and then we just watch them die -- well we really count birth and death records,  the entire cohort has to be pretty much extinct in order for us to construct a life table.  

What we do for a cohort of say 1 year olds who were born last year is different of course.  For them we create what is called a *synthetic cohort*  we do it by watching the entire population for (generally) one year.  After a year we count up all of the deaths at each age and from that we can compute the $q_x$ column of the *PERIOD* life table.  $q_x$ are the probability of death within the age category conditional on surviving to the beginning of the age. From the $q_x$ column we can compute an lx by simply starting with a big round number like 100,000 calling it $l_0$ and multiplying it by $q_0$ to get $l_1$. Then we multiply $l_1$ by $q_1$ to get $l_2$ and so on.

The details are not too important as long as you can convince yourself that it is possible to go from an array of age specific probabilities of death to number of "survivors" at each age x, $l_x$. 

But what exactly is that "period" lx that we just computed? It is the number of people who *would* survive to each age from a cohort of people whose mortality experience at each age matched the mortality experience of each age group in the population over a particular year. We call this procedure  a "synthetic" cohort approach.

Once we have the $l_x$ column for our synthetic cohort  we can produce a complete 'preiod' life table with which we can do all of the same computations as we can do with the cohort life table.  

The "period" and "cohort" life tables are different -- where the latter stresses the experience of a group of people born in the same year, the former measures the mortality experience of the entire population over a single year.  Period tables are particularly useful for discovering things, like in the news article above, that US mortality has stopped improving for two years.  That's not something we could learn from cohort life tables -- at least not about the current years.

## Using life tables to explore family level mortality experience

In what remains of this lab, we'll use the technique we developed above to generate simulated ages at death of people born in various years to estimate the probabilities of things like:  A mother born in year X outliving her child born in year Y. To do this we will want to assume that the probabilities of death among family members are *independtent*.  This is not entirely inocuous since there are  many reasons to expect that long lived parents are more likely to have long lived children:

1. Genetic proclivities toward health and illness exist.
1. Exposure to pathogens might correlate to residence.
1. Cultural habits might correlate with health.
1. Human capital of parents could affect longevity of both generations.

#### While the effects of non-independence between parent's and children's longevity will surely not be captured in the distribution (ie the spread) of our answers, will the means also be affected ?



```python
# a function to generate a simulated population
def popSim(ltab,N=100000):
    """
    uses pdeathF to add a column of probabilities of death; then draws N ages (at death) 
    """
    ltab0=pdeathF(ltab)
    ## zero probability of death is fine for ages to which no one survives -- but
    ## .sample() cannot stand it so we remove them from consideration
    ltab0=ltab0.where('pdeath',are.above(0))
 
    ltabS=ltab0.sample(k=N,with_replacement= True, weights= ltab0['pdeath'])
    aged=ltabS['Age']+ltabS['ax']
    return(aged)
```

```python
# example 0 - expectation of remaining years of life at age 15 for 1789 Sweden (cohort life tables)
# extract the 1789 life table from sweden
swe1789=sweden.where('Year',1789)
## generate a simulated array of death ages
aged=popSim(swe1789)
print(aged.mean())
swe1789.where('Age',0)
```
