## Age specific fertility rates

We can improve on the above  measures by recognizing the life cycle nature of fertility.  For biological as well as social/economic reasons, women tend to have children in their 20s and 30s and less so in either their teens or 40s.  On a population scale,  this means that populations with relatively more women in their 20s and 30s will tend to have more children then populations with older age structures *even if individuals in those populations behave identically* over their life courses.  

So how shall we characterize the age pattern of fertility?  how about by computing a specific fertility rate for each age.  Consider the following equation:

$$ASFR_a=\frac{births_a}{PYL_a}$$

where $ASFR_a$ is the age specific fertility rate of women of age $a$ over a particular period of time -- generally one year;  $births_a$ is the total number of births to all women in the population of age a (during the particular time period) And $PYL_a$  is the sum of the <i> person years lived</i> by women of age $a$ during the period over which the rate is being computed. 


The person-years lived concept is, what makes the ASFR a *rate* rather than a *probability*.  HOWEVER, in the US in the 21st Century for women of childbearing age, there is very little difference between PYL and the number of women alive at the beginning of the year.  In more difficult times (for example nearly all of human history and many poor countries today) maternal death was/is both a social and a statistical problem of heart breaking magnitude.  The former obviously, but the latter because women with higher fertility are also at greater risk of maternal death.

Fortunately in 2015 in the US PYL is quantitatively so close to  the number of women of the particular age at the beginning of the period (assuming that we are computing rates for a single year)  -- that we will simply use that figure in its place.  In other words, what we are going to compute is technically a probability -- that is numerically indistinguishable from a rate.

### Alert students will be struck by ...

The similarity of our age specific fertilitly rates with the '''period''' life table computations.  We made a big deal about the difference between *period* and *cohort* life tables... remember ?  Well this time we won't make a big deal out of it,  but note that the same issues apply. We *could* compute ASFRs for a *cohort* of women by observing those born in a particular year as they age over their entire reproductive lives. It is, however, much more common to compute *period* fertility rates as we are doing here.

To compute (period) AFSR, we'll break the problem into three steps: 
1. Divide the data into pieces according to age
1. The "numerator" will be the number of births to women of each age in the population during the year
2. The "denominator" will be the number of women at risk of giving birth at each age (times 1 year)

For each age, the ASFR will be the numerator divided by the denominator.


## The .groups() method breaks the table up according to one variable and computes something based on another.

In the cell below  we use the .groups() method to compute the sum of the fert10 variable for each subset of
the fert table defined by a different value of 'AGE'.  In other words we grab all the rows corresponding to women of age 15 and take the sum of fert10 (which turns out to be 77) then we take all the rows corresponding to 16 year old women and sum their fert10 values (133 in this case).  And we continue.  

Syntactically, the first argument of groups() is a list of the names of the columns by which we want to form subsets.  The second argument is the function, (np.nansum is just a fancy version of sum) which we want to apply to the subsets of each other variable in the table -- in this case the only other variable is fert10. 



```python
## the .groups() method divides the table 
numer=fert.select(['AGE','fert10']).groups(['AGE'],np.nansum)
numer.show(10)
```

```python
## same trick is for numer -- but we use the function len instead of np.nansum
denom=fert.select(['AGE','fert10']).groups(['AGE'],len)
denom.show(10)
```

```python
## compute the asfr and form a table for easy plotting
asfr=numer["fert10 nansum"]/denom["fert10 len"]

ASF=Table().with_column("AGE",denom['AGE'],"rate",asfr)
#ASFR.show()
ASF.scatter("AGE")
```

##  And then let's compute ASFR separately for immigrant and USborn women

But let's do it by writing a function that will compute ASFR by any categorical variable.  That will be useful later

```python
## This is kind of clumsy but the datascience module lacks a join over more than one variable 
## since we can't be sure that numer and denom are in the same order (sort only works on one variable also)
## we need to carefully divide numerator by denominator one at a time  this is really boring but so it goes.
def asfrBySomething(cvar,fdat=fert):
    ## note the addition of cvar
    numer=fdat.select(['AGE',cvar,'fert10']).groups(['AGE',cvar],np.nansum).relabel('fert10 nansum','births')
    denom=fdat.select(['AGE',cvar,'fert10']).groups(['AGE',cvar],len).relabel('fert10 len','pyl')
    ## some lists to hold results
    age=[]
    cvariable=[]
    rate=[]

    for a in np.unique(fdat['AGE']) :
        for cv in np.unique(fdat[cvar]) :
            # extract one value from the table...painfully
            num=numer.where('AGE',a).where(cvar,cv)['births']
            den=denom.where('AGE',a).where(cvar,cv)['pyl']
        
            age.append(a)
            cvariable.append(cv)
            if (len(den)>0):
                rate.append(np.asscalar(num/den))
            else:
                rate.append(0)
    # assemble the saved results into a table                
    ASFtab=Table().with_columns("AGE",age,cvar,cvariable,"rate",rate) 
            
    return(ASFtab)
```

```python
ASFimmig=asfrBySomething("immig",fdat=fert)
import pylab
pylab.plot(ASFimmig.where('immig',0)['AGE'],ASFimmig.where('immig',0)['rate'],label='US Born')
pylab.plot(ASFimmig.where('immig',1)['AGE'],ASFimmig.where('immig',1)['rate'],label='immigrant')
pylab.title("ASFR US Born and immigrant women 2015")
pylab.legend(loc='upper right')
```

## The Total Fertility Rate 

The ASFR table that we computed for immigrants and US born women is nice. It shows that both groups have roughly similarly shaped age patterns of fertility--which is nice to know -- but it's not  the question that we (or Francis Walker) were asking. What we really want is a summary measure  that *takes away the effect of age*.   From the  crude measure that we calculated for immigrant and US born women way back in cell called "Crude fertility rates" we learned that on average immigrant women had more babies than US born women in 2015 -- but that measure was weak because immigrant and US born women might  significantly different age distributions and that could obviously affect the "crude" rates by quite a lot. 

The graph above doesn't quite get us off the hook.  It was worth a try because we *might* have gotten lucky and found that the fertility rates for one groups were  higher or lower at **all** ages, but we didn't find that -- and consequently, we were reduced to shrugging and looking sheepishly away. 

To avoid this sort of humiliation, demographers have devised the Total Fertility Rate (TFR).

The TFR is often **misleadingly** referred to as "the number of children per woman".  But just like "life expectancy"  that we encountered last week,  it's more complicated than it seems. Also like life expectancy,  there is both a cohort and a period version. We are only looking at the period version today since cohort data on fertility are quite rare.

Once we have computed the age specific fertility rates,  TFR is quite easy to compute.

```python
# TFR stands for 'total fertility rate'  it is easy to compute once
# we have an array of age specific fertility rates
tfr=np.sum(ASF['rate'])
tfr
```

## Decomposition 

In the case of TFRs and many but not all statistics that we might compute, it is possible to view the TFR of the entire population as a **weighted average** of the TFR of its sub-groups.  That is what we sort of showed in the "Decomposing TFR" cell.  There is an unfortunate amount of rounding error in this case which appear to be exacerbated by the rather small rates at the lower ages.  If you care to investigate, you'll find that the error is much reduced if we include only 18+ year olds in the computations.


Suppose we compute TFRs for each combination of educ2 and immig (6 TFRs in all) and then we compute the *education adjusted TFR* for immigrants as:

$$ TFR_{\text{adjusted}} = \sum_{\text{x in educ2}} Prop_{\text{US born with educ2 = x}}*TFR_{\text{immigrants with educ2 =x}} $$

The notation is clumsy but the idea is simple -- as we did in the cell called "Decomposing TFR" we are just multiplying the proportion of a population by a TFR since the proportions sum to 1, it's a weighted avearge.

The insight in the equation above is that we can use the **proportions of one group and the TFRs of the other**

That would directly (sort of) answer the question : what if immigrant women had the same education levels as US born women.
